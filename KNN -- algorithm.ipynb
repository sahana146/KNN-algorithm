```python
import csv
import math
import operator
import random

def splitdata(filename,train,test):                      #dividing the dataset into training and testing set
    with open(filename,'r') as file:
        data=csv.reader(file)
        datalist=list(data)
        for i in range(len(datalist)-1):
            for j in range(4):
                datalist[i][j]=float(datalist[i][j])
                
            if random.random()<0.66:
                train.append(datalist[i])
            else:
                test.append(datalist[i])    

                
def caldistance(train,testcase,distance):                      #calculating euclidean distance
    for i in range(len(train)-1):
        euclideandist=0
        for j in range(4):
            euclideandist+=math.pow((train[i][j]-testcase[j]),2)      
        euclideandist=math.sqrt(euclideandist)
        distance.append((train[i][4],euclideandist))
 

        
def neighbour(distance,k,neighbours):                          #determining the closest neighbours based on distance
    distance.sort(key=operator.itemgetter(1))
    for i in range(k):
        neighbours.append(distance[i][0])
    
    
    

def main():                                             #main function
    train=[]
    test=[]
    resultobtained=[]
    k=3                                                 #taking K=3 (to achieve higher accuracy)#
                                                        #k usually ranges from 1 to 21
    filename="iris.data.txt"                            
    splitdata(filename,train,test)                      
    for j in range(len(test)-1):
        distance=[]
        caldistance(train,test[j],distance)
        neighbours=[]
        neighbour(distance,k,neighbours)
        classvalue={}
        for i in range(k):                                 #determining the flower species  
                                                           #by finding the number of occurences of each specie 
                                                           #among the neighbours
            if neighbours[i] in classvalue:
                classvalue[neighbours[i]]+=1
            else:
                classvalue[neighbours[i]]=1 
        classvalue=sorted(classvalue.items(),key=operator.itemgetter(1),reverse=True)
        print(test[j])
        print(classvalue[0][0])                            #printing the flower specie
        resultobtained.append(classvalue[0][0])            #storing the result obtained
    correct=0
    for x in range(len(test)-1):                           #calculating the accuracy by 
                                                           #comparing expected outcome with obtained outcome
        if resultobtained[x]==test[x][-1]:
            correct+=1
            
    print("\naccuracy "+str((correct/len(test))*100)+'%'+'\n')
    
    
if __name__ =='__main__':
    main()                                                #calling the main function
    
```
