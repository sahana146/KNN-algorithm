```python
import numpy
import pandas
import math
import operator
import random

def splitdata(datalist,train,test,col_list):                      #dividing the dataset into training and testing set
    train.append(datalist[0])
    test.append(datalist[0])
    for i in range(1,len(datalist)-1):
        if random.random()<0.75:
            train.append(datalist[i])
        else:
            test.append(datalist[i])    

                
def caldistance(train,testcase,distance,col_list):                      #calculating euclidean distance
    for i in range(1,len(train)-1):
        euclideandist=0
        for j in col_list:
            euclideandist+=math.pow(float(float(train[i][j])-float(testcase[j])),2)      
        euclideandist=math.sqrt(euclideandist)
        distance.append((train[i][31],euclideandist))
 

        
def neighbour(distance,k,neighbours):                          #determining the closest neighbours based on distance
    distance.sort(key=operator.itemgetter(1))
    for i in range(k):
        neighbours.append(distance[i][0])
    
    
    

def main():                                             #main function
    train=[]
    test=[]
    resultobtained=[]
                                                    
    filename="players_stats.csv" 
    csv = numpy.genfromtxt(filename,dtype=None,delimiter=',',usecols=numpy.arange(0,34),encoding="ascii")
    
    k=10
   
    
    col_list=[6,9,12,13,14,15,16,17,18,19,20]                      #feature columns
                                                                #class column--->position
                                                              
    delete=[]
    for j in range(1,len(csv[:,31])-1):                         #indexes of all rows with missing value
            if csv[j,31]=='':
                delete.append(j)
             
                                                        
    csv=numpy.delete(csv,delete,0)                       #handling missing value by deleting the row with missing value
   
    
    for i in col_list:                                          # normalization
        mi=float(min(csv[1:,i]))
        ma=float(max(csv[1:,i]))
        for j in range(1,len(csv)-1): 
            csv[j,i]=(float(csv[j,i])-mi)/(ma-mi)               #standard normalization formula
                
    splitdata(csv,train,test,col_list)        
    
    
    for j in range(1,len(test)-1):
        distance=[]
        caldistance(train,test[j],distance,col_list)
        
        neighbours=[]
        neighbour(distance,k,neighbours)
        print("Neighbours \n",neighbours)
        classvalue={}
        for i in range(k):                                 #determining the position  
                                                           #by finding the number of occurences of each position 
                                                           #among the neighbours
            if neighbours[i] in classvalue:
                classvalue[neighbours[i]]+=1
            else:
                classvalue[neighbours[i]]=1 
        classvalue=sorted(classvalue.items(),key=operator.itemgetter(1),reverse=True)
        print("testcase \n",test[j])
        print("output \n",classvalue[0][0])                            #printing the position
        resultobtained.append(classvalue[0][0])            #storing the result obtained
    correct=0
    
    for x in range(1,len(test)-1):                           #calculating the accuracy by 
                                                           #comparing expected outcome with obtained outcome
        if resultobtained[x-1]==test[x][31]:
            correct+=1
            
    print("\naccuracy "+str((correct/len(test))*100)+'%'+'\n')
    
    
if __name__ =='__main__':
    main()                                                #calling the main function
    
```
